```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'http://www.imdb.com/chart/top'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python
movies = soup.select('td.titleColumn')
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value')
        for b in soup.select('td.posterColumn span[name=ir]')]
```

```python
list = []
 
# Iterating over movies to extract
# each movie's details
for index in range(0, len(movies)):
     
    # Separating movie into: 'place',
    # 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    place = movie[:len(str(index))-(len(movie))]
    data = {"place": place,
            "movie_title": movie_title,
            "rating": ratings[index],
            "year": year,
            "star_cast": crew[index],
            }
    list.append(data)
print(list)
```

```python
#saving the list as dataframe
df = pd.DataFrame(list)
df
```

```python
list_ids=[]
for line in range(len(tables)):
    #a=tables[line].find_all("a")
    for l in tables[line].find_all("a"):
        id=re.findall("tt[\w]+", l["href"])
        list_ids.append(id[0])

list_ids
```

```python
import requests

url = "https://movie-database-alternative.p.rapidapi.com/"
list_response=[]
for id in list_ids:
    querystring = {"r":"json","i":id}

    headers = {
        "X-RapidAPI-Key": "2a936c3ee2msh60bb07688a7e74ap12f566jsnd92e5c58d201",
        "X-RapidAPI-Host": "movie-database-alternative.p.rapidapi.com"
    }

    response = requests.request("GET", url, headers=headers, params=querystring)
    list_response.append(response.json())
    print(list_response)
```
